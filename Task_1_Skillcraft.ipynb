```python
import pandas as pd
import matplotlib.pyplot as plt

# File paths
file_country_metadata = '/content/Metadata_Country_API_SP.POP.TOTL_DS2_en_csv_v2_900.csv'

# Load the country metadata dataset
country_metadata_df = pd.read_csv(file_country_metadata)

# Distribution of countries by Income Group
income_group_counts = country_metadata_df['IncomeGroup'].value_counts().dropna()

# Plotting the distribution
plt.figure(figsize=(10, 6))
income_group_counts.plot(kind='bar', color='skyblue', edgecolor='black')

# Customizing the plot
plt.title('Distribution of Countries by Income Group')
plt.xlabel('Income Group')
plt.ylabel('Number of Countries')
plt.xticks(rotation=45)
plt.grid(axis='y', linestyle='--', alpha=0.7)

# Display the plot
plt.show()
```
